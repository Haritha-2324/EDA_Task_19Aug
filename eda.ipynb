KPIs & Metrics

```python
# Example calculations to inform KPIs (don't save these; just use for planning)
if 'Revenue' in df.columns:
    total_revenue = df['Revenue'].sum()
    print("Total Revenue:", total_revenue)

if 'Order_Value' in df.columns and 'Order_ID' in df.columns:  # Assuming relevant columns
    avg_order_value = df['Order_Value'].mean()
    print("Average Order Value:", avg_order_value)

# Revenue: Product Category (group by)
if 'Revenue' in df.columns and 'Product_Category' in df.columns:
    revenue_by_category = df.groupby('Product_Category')['Revenue'].sum()
    print("\nRevenue by Product Category:\n", revenue_by_category)

# Sales: Trend Over Time
if 'Date' in df.columns and 'Revenue' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])  # Ensure date type
    monthly_trend = df.resample('M', on='Date')['Revenue'].sum()
    print("\nMonthly Revenue Trend:\n", monthly_trend)

```
